```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train=pd.read_csv(r'C:\Users\Lenovo\Desktop\kaggle\blackfriday_train.csv')
df_train.head()
```

```python
df_test=pd.read_csv(r"C:\Users\Lenovo\Desktop\kaggle\blackfriday_test.csv")
df_test.head()
```

```python
df=df_train.append(df_test)
df.head()
```

```python
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#handling the categorical feature genders

df['Gender']=df['Gender'].map({'F':0,'M':1})
```

```python
df.head()
```

```python
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df.head()
```

```python
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
df_city.head()
```

```python
df1=pd.concat([df,df_city],axis=1)
df1
```

```python
df1.drop('City_Category',axis=1,inplace=True)
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1['Product_Category_2'].unique()
```

```python
df1['Product_Category_2'].value_counts()
```

```python
a=df1['Product_Category_2'].mode()[0]
```

```python
#replacing the missing value with mode

df1['Product_Category_2']=df1['Product_Category_2'].fillna(a)
df1['Product_Category_2'].isnull().sum()
```

```python
#replacing the missing value of df1['Product_Category_3'] with mode
```

```python
df1['Product_Category_3'].unique()
```

```python
b=df1['Product_Category_3'].mode()[0]
b
```

```python
df1['Product_Category_3']=df1['Product_Category_3'].fillna(b)
df1['Product_Category_3'].isnull().sum()
```

```python
df1.head()
```

```python
df1['Stay_In_Current_City_Years']=df1['Stay_In_Current_City_Years'].str.replace('+','')
df1.head()
```

```python
#converting the object into the integer
```

```python
df1['Stay_In_Current_City_Years']=df1['Stay_In_Current_City_Years'].astype(int)
df1.info()
```

```python
df1['B']=df1['B'].astype(int)
df1['C']=df1['C'].astype(int)
```
