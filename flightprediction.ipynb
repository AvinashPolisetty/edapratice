```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train=pd.read_excel('Data_Train.xlsx')
df_train.head()
```

```python
df_test=pd.read_excel("Test_set.xlsx")
df_test.head()
```

```python
df_final=df_train.append(df_test)
df_final.head()
```

```python
df_final.info()
```

```python
#convert the data  into integers
df_final['date']=df_final['Date_of_Journey'].str.split('/').str[0]
df_final['month']=df_final['Date_of_Journey'].str.split('/').str[1]
df_final['year']=df_final['Date_of_Journey'].str.split('/').str[2]
df_final.head(3)
```

```python
df_final['date']=df_final['date'].astype(int)
df_final['month']=df_final['month'].astype(int)
df_final['year']=df_final['year'].astype(int)
```

```python
df_final.info()
```

```python
df_final.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df_final
```

```python
df_final['Arrival_Time']=df_final['Arrival_Time'].str.split(' ').str[0]
```

```python
df_final
```

```python
df_final['atime_hours']=df_final["Arrival_Time"].apply(lambda x:int(x.split(" ")[0].split(":")[0]))
df_final['atime_mins']=df_final["Arrival_Time"].apply(lambda x:int(x.split(" ")[0].split(":")[1]))
```

```python
df_final['atime_hours']=df_final['atime_hours'].astype(int)
df_final['atime_mins']=df_final['atime_mins'].astype(int)
```

```python
df_final.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df_final
```

```python
df_final['derhours']=df_final['Dep_Time'].str.split(':').str[0]
df_final['depmints']=df_final['Dep_Time'].str.split(':').str[1]
```

```python
df_final['derhours']=df_final['derhours'].astype(int)
df_final['depmints']=df_final['depmints'].astype(int)
```

```python
df_final.drop('Dep_Time',axis=1,inplace=True)
```

```python
df_final.head()
```

```python
df_final['Total_Stops']=df_final['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
df_final.drop('Route',axis=1,inplace=True)
```

```python
df_final
```

```python
df_final['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
df_final['Airline']=labelencoder.fit_transform(df_final['Airline'])
df_final['Source']=labelencoder.fit_transform(df_final['Source'])
df_final['Destination']=labelencoder.fit_transform(df_final['Destination'])
df_final['Additional_Info']=labelencoder.fit_transform(df_final['Additional_Info'])
                                                                                              
                                            
```
